```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
```

```python
def ScrapS7():
    list_appended=[]
    for i in range(37):
        urltempo=urlopen("https://www.amazon.fr/Samsung-Galaxy-S7-Edge-Smartphone/product-reviews/B01BYD0TZM/ref=cm_cr_othr_d_paging_btm_1?ie=UTF8&reviewerType=all_reviews&pageNumber={}".format(i+1))

        souptempo = BeautifulSoup(urltempo, 'lxml')
        list_scraped = scrapfull(souptempo)
        list_appended.extend(list_scraped)
    return(list_appended)
```

```python
def scrapfull(soup):
    emptylist=[]
    for i in soup.find_all('span', {"class":"a-size-base review-text" }):
        emptylist.append(i.text)
    for j in soup.find_all("a", {"class":"a-size-base a-link-normal review-title a-color-base a-text-bold"}):
        emptylist.append(j.text)
        return(emptylist)
```
